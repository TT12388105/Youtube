```python
# importing modules
# pip install requests

import requests
from datetime import date
```

```python
# url

url = 'http://astrometry.ch:5000/perihelion/'
```

```python
# making request

r = requests.get(url)
data = r.json()
data
```

```python
# getting all listed bodies

all_bodies = []

for item in data:
    all_bodies.append(list(item.keys())[1])
    
all_bodies
```

```python
# getting data for specific body

body = 'Mars'

for item in data:
    if body in item.keys():
        peri = item[body]
        break
        
peri
```

```python
# printing all perihelion dates

for i in peri:
    d = list(map(int, i.split(',')))
    dt = date(d[0],d[1],d[2])
    print(dt.strftime('%d %b, %Y'))
```
